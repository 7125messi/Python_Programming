## Use of "group_by" in Pandas

```python
import pandas as pd
```

```python
# Here we'll make dataset from alcohol consumption by country. 
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
# to see some of the data
drinks.head()
```

```python
# average beer servings: 
drinks.beer_servings.mean()
```

```python
# to get beer servings by 'continent'
drinks.groupby('continent')['beer_servings'].mean()
```

### Applying Conditions

```python
# to get data about 'Africa Continent'
drinks[drinks.continent == 'Africa']
```

### TO get the mean about beer_servings:

```python
drinks[drinks.continent == 'Africa'].beer_servings.mean()
```

### Or we can do this below: 

```python
drinks[drinks.continent == 'Africa']['beer_servings'].mean()
```

### To get maximum beer servings for each continent. 

```python
# we can use other aggregate functions too:
drinks.groupby('continent')['beer_servings'].max()
```

### use of agg()

```python
drinks.groupby('continent').beer_servings.agg(['count','min','max','mean'])
```
