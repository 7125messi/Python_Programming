```python
import matplotlib.pyplot as plt
%matplotlib inline   # this magic function is included so that our graph could render fine. 
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather Data')
plt.plot(x,y, color = 'green', linewidth =4, linestyle = 'dotted')
```

## Format String in plot funcion

```python
plt.plot(x,y, 'r+-.')  
```

```python
# we can do so by supplying different arguments also, it's same as the previous one
plt.plot(x,y, color = 'red', marker='D', linestyle='--', markersize=10)
```

## alpha funcion for transparency

```python
plt.plot(x,y, color = 'red', marker='D', linestyle='--', markersize=10, alpha = 0.6)
```

```python
plt.plot([1,2,3], [1,2,3], 'go-', label='line 1', linewidth=1)
plt.plot([1,2,3], [1,4,9], 'rs-',  label='line 2')
# follow this link for more parameters: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html
```

## Axes labels, Legends, Grid

```python
# Weather data:
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
# plotting a graph for maximum temperature
plt.plot(days,max_t)
```

```python
# we can do max, min and average all in one chart
plt.xlabel('Days')  # for x - axis label
plt.ylabel('Temperature')  # for y - axis label
plt.title('Weather Report')  # for title of the graph
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

### Legends and Grids

```python
plt.xlabel('Days')  # for x - axis label
plt.ylabel('Temperature')  # for y - axis label
plt.title('Weather Report')  # for title of the graph
plt.plot(days,max_t, label= 'Max')
plt.plot(days,min_t, label = 'Min')
plt.plot(days,avg_t, label = 'Avg')
# to appear legend
plt.legend(loc = 'best', shadow = True, title = 'Weather Report')   # default is best
# we can set the grid too
plt.grid()
```
