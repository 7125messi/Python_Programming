```python
import pandas as pd
```

```python
df1 =pd.DataFrame( {
    'city' : ['new york', 'chicago', 'orlando'],
    'temperature' : [21,14,35]    
})
df1
```

```python

df2 =pd.DataFrame( {
    'city' : ['chicago', 'new york', 'orlando'],
    'humidity' : [65,68,75]    
})
df2
```

```python
# to merge data frames
df3 = pd.merge(df1,df2,on='city')
df3
# this is exactly same as databse join
# this acts like inner join 
```

### Merging Un-Usual Columns and Outer join, left and right outer join

```python
df4 =pd.DataFrame( {
    'city' : ['new york', 'chicago', 'orlando','baltino'],
    'temperature' : [21,14,35,27]    
})
df4
```

```python
df5 =pd.DataFrame( {
    'city' : ['chicago', 'new york', 'san fransisco'],
    'humidity' : [65,68,75]    
})
df5
```

```python
df6 = pd.merge(df4,df5,on='city',how='inner')  # inner is default
df6
```

```python
# left outer join
df7 = pd.merge(df4,df5,on='city',how='left')
df7
```

```python
# right outer join
df8 = pd.merge(df4,df5,on='city',how='right')
df8
```

```python
# full outer join
df9 = pd.merge(df4,df5,on='city',how='outer')
df9
```

```python
# to know from which table data came, we can do it by indicator option
df10 = pd.merge(df4,df5,on='city',how='outer', indicator=True)
df10
```

### Use of suffixes

```python
df11 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df11
```

```python
df12 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df12
```

```python
# if we merge simply
df13 = pd.merge(df11,df12, on = 'city', how = 'outer')
df13
```

```python
# if we merge with suffixes
df13 = pd.merge(df11,df12, on = 'city', how = 'outer', suffixes=('_first','_second'))
df13
```

### Use of set_index()  and  lsuffix()  and  rsuffix()

```python
df1 = pd.DataFrame({
    'city':['new york','chicago','orlando'],
    'temperature':[21,14,35]
})
df1.set_index('city', inplace=True)
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2.set_index('city', inplace=True)
df2
```

```python
df1.join(df2, lsuffix='_l', rsuffix='_r')
```
